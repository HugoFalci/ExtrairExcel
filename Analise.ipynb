```python
import pandas as pd
```

```python
df_path = '1449.xlsx'
df = pd.read_excel(df_path)
```

```python
print(df.head)
```

```python
coluna_df = df[['Inquilino Nome', 'Inquilino CPF/CNPJ', 'Inquilino Celular', 'Inquilino E-mail', 'Proprietario Nome',
                'Proprietario CPF/CNPJ', 'Proprietario Celular', 'Proprietario E-mail', 'Imovel Nome', 'Imovel CEP', 
                'Imovel Numero', 'Imovel Complemento', 'valor_vigente', 'data_inicio_contrato', 'duracao', 
                'tem_comissao_inicial','porcentagem_comissao_inicial', 'valor_comissao_inicial', 'tem_taxa_administracao', 
                'porcentagem_taxa_administracao', 'valor_taxa_administracao', 'primeiro_vencimento', 
                'ultimo_vencimento_pago']]
```

```python
def tratar_data(value):
    if pd.isna(value) or value == '':
        return pd.NA
    try:
        return pd.to_datetime(value, dayfirst=True).strftime('%d/%m/%Y')
    except Exception:
        return pd.NA
```

```python
coluna_df.loc[:, 'data_inicio_contrato'] = coluna_df['data_inicio_contrato'].apply(tratar_data)
coluna_df.loc[:, 'primeiro_vencimento'] = coluna_df['primeiro_vencimento'].apply(tratar_data)
coluna_df.loc[:, 'ultimo_vencimento_pago'] = coluna_df['ultimo_vencimento_pago'].apply(tratar_data)
```

```python
coluna_df['Proprietario CPF/CNPJ'] = coluna_df['Proprietario CPF/CNPJ'].str.strip()
coluna_df['Inquilino CPF/CNPJ'] = coluna_df['Inquilino CPF/CNPJ'].str.strip()
coluna_df.dropna(subset=['data_inicio_contrato', 'primeiro_vencimento', 'ultimo_vencimento_pago'], inplace=True)
```

```python
nome_planilha = 'Resultado.xlsx'
coluna_df.to_excel(nome_planilha, index=False)
```
